```python
import numpy as np
import pandas as pd
```

```python
x=pd.read_excel('keystroke.xls')
```

```python
print(x.columns)
```

```python
del x['sessionIndex']
del x['rep']
```

```python
print(x.columns)
```

```python
x.describe(include='all')
```

```python
x1=x.groupby('subject')
subname=x1.subject.unique()
print(subname)
```

```python
for i in range(len(subname)):
    p=subname.iloc[i]
    print(x1.get_group(p[0]).describe(include='all'))
```

```python
import numpy as np
import scipy.stats as stats
import pylab as pl
p=subname.iloc[0]
balu=x1.get_group(p[0])
balu=sorted(balu.iloc[:,1])
fit = stats.norm.pdf(balu, np.mean(balu), np.std(balu))  #his is a fitting indeed

pl.figure
pl.plot(balu,fit,'-o')

pl.hist(balu,normed=True)      #use this to draw histogram of your data

pl.show()
```
